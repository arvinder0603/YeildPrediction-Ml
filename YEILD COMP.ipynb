```python
import numpy as np
from scipy import stats

# Define the threshold for outlier detection
threshold = 3

# Iterate over each column in the DataFrame
for column in df.columns:
    # Calculate the Z-scores for each data point in the column
    z_scores = np.abs(stats.zscore(df[column]))

    # Identify outlier rows based on the threshold
    outlier_mask = z_scores > threshold

    # Remove outliers from the column
    df = df[~outlier_mask]

# Reset the indices of the DataFrame
df = df.reset_index(drop=True)

```

# GRADIENT BOOSTING

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
n_estimators = 100  # Number of boosting stages
learning_rate = 0.1  # Learning rate (shrinks the contribution of each tree)
max_depth = 3
```

```python
model = GradientBoostingRegressor(n_estimators=n_estimators, learning_rate=learning_rate, max_depth=max_depth)
```

```python

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(y_test, y_pred)

mae = mean_absolute_error(y_test, y_pred)
# Print the evaluation metrics
print("Mean Squared Error:", mse)

print("Mean Absolute Error:", mae)
```
